```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import utils

from sklearn import tree

np.random.seed(0)
```

# The spam email dataset

```python
emails = np.array([
    [7,8,1],
    [3,2,0],
    [8,4,1],
    [2,6,0],
    [6,5,1],
    [9,6,1],
    [8,5,0],
    [7,1,0],
    [1,9,1],
    [4,7,0],
    [1,3,0],
    [3,10,1],
    [2,2,1],
    [9,3,0],
    [5,3,0],
    [10,1,0],
    [5,9,1],
    [10,8,1],
])
spam_dataset = pd.DataFrame(data=emails, columns=["Lottery", "Sale", "Spam"])
spam_dataset
```

```python
features = spam_dataset[['Lottery', 'Sale']]
labels = spam_dataset['Spam']
utils.plot_points(features, labels)
```

# AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
# Set the random_state so that we always get the same results
# use adaboost classifier with 6 estimator
adaboost_classifier = AdaBoostClassifier(n_estimators=6).fit(features, labels)

adaboost_classifier.score(features, labels)
```

```python
utils.plot_model(features, labels, adaboost_classifier)
```

```python
estimators = adaboost_classifier.estimators_
for estimator in estimators:
    utils.plot_model(features, labels, estimator)
    plt.show()
```

```python
adaboost_classifier = AdaBoostClassifier(n_estimators=10).fit(features, labels)

adaboost_classifier.score(features, labels)
utils.plot_model(features, labels, adaboost_classifier)
```

```python
estimators = adaboost_classifier.estimators_
for estimator in estimators:
    utils.plot_model(features, labels, estimator)
    plt.show()
```

```python
adaboost_classifier = AdaBoostClassifier(n_estimators=14).fit(features, labels)

adaboost_classifier.score(features, labels)
utils.plot_model(features, labels, adaboost_classifier)
estimators = adaboost_classifier.estimators_
for estimator in estimators:
    utils.plot_model(features, labels, estimator)
    plt.show()
```
